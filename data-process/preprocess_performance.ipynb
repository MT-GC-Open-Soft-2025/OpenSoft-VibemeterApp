```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_regression, f_classif
from scipy.stats import pearsonr
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('/content/performance_dataset.csv')
```

```python
df.head()
```

```python
df = df.drop('Review_Period', axis=1)
```

```python
df.columns
```

```python
df.sort_values(by=["Employee_ID"], ascending=[True], inplace=True)
df.head()
```

```python
df['Manager_Feedback'].unique()
```

```python
df['Manager_Feedback'] = df['Manager_Feedback'].replace(['Needs Improvement'], 0)
df['Manager_Feedback'] = df['Manager_Feedback'].replace(['Meets Expectations'], 1)
df['Manager_Feedback'] = df['Manager_Feedback'].replace(['Exceeds Expectations'], 2)
df['Promotion_Consideration'] = df['Promotion_Consideration'].replace(True, 1)
df['Promotion_Consideration'] = df['Promotion_Consideration'].replace(False, 0)
df
df.head()
```

```python
combined_df = df.groupby("Employee_ID")[['Manager_Feedback', 'Promotion_Consideration','Performance_Rating']].mean().reset_index()

combined_df
```

```python
num_unique_employees = combined_df['Employee_ID'].nunique()
num_rows = combined_df.shape[0]

if num_unique_employees == num_rows:
  print("The number of unique employee IDs is equal to the number of rows.")
else:
  print("The number of unique employee IDs is not equal to the number of rows.")
```

```python
output_path = "unique_employees_weighted_performance.csv"
combined_df.to_csv(output_path, index=False)

print(f"Processed data saved to {output_path}")
```
