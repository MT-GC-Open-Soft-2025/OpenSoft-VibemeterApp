```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('onboarding_dataset.csv')
df.head()
```

```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'], errors='coerce')

df['Mentor_Assigned'] = df['Mentor_Assigned'].astype(str)
df['Initial_Training_Completed'] = df['Initial_Training_Completed'].astype(str)

df['Mentor_Assigned'] = df['Mentor_Assigned'].map({'False': 0, 'True': 1})
df['Onboarding_Feedback'] = df['Onboarding_Feedback'].map({'Poor': 0, 'Average': 1, 'Good': 2, 'Excellent': 3})
df['Initial_Training_Completed'] = df['Initial_Training_Completed'].map({'False': 0, 'True': 1})

df_sorted = df.sort_values(by=['Employee_ID', 'Joining_Date'], ascending=[True, False])
df_unique = df_sorted.drop_duplicates(subset=['Employee_ID'], keep='first')

print(df_unique.head())


df_unique.to_csv('processed_onboarding_dataset.csv', index=False)

```
