```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('leave_dataset.csv')
```

```python
df.head()
```

```python


df.drop(columns=['Leave_Start_Date'], inplace=True)

df['Leave_End_Date'] = pd.to_datetime(df['Leave_End_Date'], errors='coerce')

df.sort_values(by=['Employee_ID', 'Leave_End_Date'], ascending=[True, False], inplace=True)



df['Leave_Type'] = df['Leave_Type'].astype(str).str.strip().str.lower()
leave_type_mapping = {
    'unpaid leave': 0,
    'sick leave': 1,
    'casual leave': 2,
    'annual leave': 3
}


df['Leave_Type'] = df['Leave_Type'].map(leave_type_mapping).astype(int)

df.head()
```

```python
df_unique = df.groupby('Employee_ID', as_index=False).agg({
    'Leave_Type': 'mean', 
    'Leave_Days': 'sum',  
    'Leave_End_Date': 'max'  
})

df_unique.head()
```
