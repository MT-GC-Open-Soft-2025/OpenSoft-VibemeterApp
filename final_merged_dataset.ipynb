```python
new_activity = activity.groupby('Employee_ID')['Work_Hours'].sum().reset_index()

new_activity.rename(columns={'Work_Hours': 'total_work_hours'}, inplace=True)

min_work_hours = new_activity['total_work_hours'].min()

new_activity['total_work_hours'].fillna(min_work_hours, inplace=True)

new_activity
```

```python
new_leave = leave.groupby('Employee_ID')['Leave_Type'].value_counts().unstack(fill_value=0).reset_index()
new_leave['types_of_leaves'] = new_leave.apply(lambda row: dict(zip(row.index[1:], row.values[1:])), axis=1)
new_leave = new_leave[['Employee_ID', 'types_of_leaves']]
new_leave.head()
```

```python
unique_employee_ids = set()

for df in [df_activity, df_performance, df_rewards, df_leave, activity, performance, rewards, leave, on_boarding]:
  unique_employee_ids.update(df['Employee_ID'].unique())

print(len(unique_employee_ids))
```

```python
employee_dataframes = []

for employee_id in unique_employee_ids:
    activity_row = df_activity[df_activity['Employee_ID'] == employee_id]
    rewards_row = df_rewards[df_rewards['Employee_ID'] == employee_id]
    performance_row = df_performance[df_performance['Employee_ID'] == employee_id]
    leave_row = df_leave[df_leave['Employee_ID'] == employee_id]
    new_activity_row = new_activity[new_activity['Employee_ID'] == employee_id]
    new_reward_row = new_reward[new_reward['Employee_ID'] == employee_id]
    new_leave_row = new_leave[new_leave['Employee_ID'] == employee_id]
    new_onboarding_row = new_onboarding[new_onboarding['Employee_ID'] == employee_id]

    employee_dict = {
        'Employee_ID': employee_id,
        'should_chat_activity': None,
        'should_chat_rewards': None,
        'should_chat_performance': None,
        'should_chat_leave': None,
        'Weighted_vibe_Score': None,
        'total_work_hours': None,
        'Leave_Days': None,
        'types_of_leaves': None,
        'feedback': None,
        'weighted_performance':None,
        'reward_points': None,
        'award_list': None
    }

    if not activity_row.empty:
        employee_dict['should_chat_activity'] = activity_row['should_chat_activity'].iloc[0]
        employee_dict['Weighted_vibe_Score'] = activity_row['Weighted_vibe_Score'].iloc[0]
    if not rewards_row.empty:
        employee_dict['should_chat_rewards'] = rewards_row['should_chat_rewards'].iloc[0]
    if not performance_row.empty:
        employee_dict['should_chat_performance'] = performance_row['should_chat_performance'].iloc[0]
    if not leave_row.empty:
        employee_dict['should_chat_leave'] = leave_row['should_chat_performance'].iloc[0]
        employee_dict['Leave_Days'] = leave_row['Leave_Days'].iloc[0]
    if not new_activity_row.empty:
        employee_dict['total_work_hours'] = new_activity_row['total_work_hours'].iloc[0]
    if not new_leave_row.empty:
        employee_dict['types_of_leaves'] = new_leave_row['types_of_leaves'].iloc[0]
    if not new_onboarding_row.empty:
        employee_dict['feedback'] = new_onboarding_row['feedback'].iloc[0]
    if not performance_row.empty:
        employee_dict['weighted_performance'] = performance_row['combined_feature_performance'].iloc[0]
    if not new_reward_row.empty:
        employee_dict['reward_points'] = new_reward_row['Rewards_Points'].iloc[0]
        employee_dict['award_list'] = new_reward_row['award_list'].iloc[0]

    employee_df = pd.DataFrame([employee_dict])

    employee_dataframes.append(employee_df)


new_df = pd.concat(employee_dataframes, ignore_index=True)

print(new_df.shape)
new_df.head()
```

```python
default_dict = {'Annual Leave': 0, 'Sick Leave': 0, 'Casual Leave': 0, 'Unpaid Leave': 0}

new_df['types_of_leaves'] = new_df['types_of_leaves'].apply(lambda x: default_dict if pd.isna(x) else x)
```

```python
new_df['reward_points'].fillna(0, inplace=True)
new_df['weighted_performance'].fillna(0, inplace=True)
new_df['Leave_Days'].fillna(0, inplace=True)
new_df['total_work_hours'].fillna(new_df['total_work_hours'].min(), inplace=True)
new_df['Weighted_vibe_Score'].fillna(-1, inplace=True)
new_df['award_list'] = new_df['award_list'].apply(lambda x: x if x else [])
```

```python
columns_to_sort = [
    "should_chat_activity",
    "should_chat_rewards",
    "should_chat_performance",
    "should_chat_leave",
]

def get_sorted_factors(row):
    valid_factors = {
        col.replace("should_chat_", ""): row[col]
        for col in columns_to_sort if pd.notna(row[col])
    }

    sorted_factors = sorted(valid_factors.keys(), key=lambda x: valid_factors[x], reverse=True)

    return [str(factor) for factor in sorted_factors]

new_df["factors_in_sorted_order"] = new_df.apply(get_sorted_factors, axis=1)

new_df[["Employee_ID", "factors_in_sorted_order"]].head()
```

```python
columns_to_drop = [
    "should_chat_activity",
    "should_chat_rewards",
    "should_chat_performance",
    "should_chat_leave",
]
final_df = new_df.drop(columns=columns_to_drop)
```

```python
cols = list(final_df.columns)
cols.insert(1, cols.pop(cols.index('factors_in_sorted_order')))
final_df = new_df[cols]

final_df
```
